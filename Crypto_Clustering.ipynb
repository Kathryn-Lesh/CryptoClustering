```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
std_scaler = StandardScaler()
```

```python
# Create a DataFrame with the scaled data
df_market_data_scaled = std_scaler.fit_transform(df_market_data.to_numpy())
df_market_data_scaled = pd.DataFrame(df_market_data_scaled)

df_market_data_scaled.columns = df_market_data.columns
df_market_data_scaled.index = df_market_data.index
#df_market_data_scaled["coin"] = df_market_data.index
df_market_data_scaled.head()
```

---

### Find the Best Value for k Using the Original Data.

```python
k = list(range(1,11))
inertia = []

for i in k:
    k_model = KMeans(n_clusters = i, random_state=1)
    k_model.fit(df_market_data_scaled[['price_change_percentage_24h', 'price_change_percentage_7d']])
    inertia.append(k_model.inertia_)
```

```python
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow
```

```python
# Plot the inertias
elbow1 = df_elbow.hvplot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve", 
    xticks=k
)
elbow1
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** It looks like the best value for `k` is 4, which is just past the elbow, although I could be convinced to use k=3.

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Create model and predict clusters for k=4
model = KMeans(n_clusters=4, random_state=1)
model.fit(df_market_data_scaled[['price_change_percentage_24h', 
                                 'price_change_percentage_7d']])
predictions = model.predict(df_market_data_scaled[['price_change_percentage_24h', 
                                                   'price_change_percentage_7d']])
predictions
```

```python
df_market_data_scaled_copy = df_market_data_scaled.copy()
df_market_data_scaled_copy['clusters_4'] = predictions
df_market_data_scaled_copy.head()
```

```python
# Plot the clusters
cluster_plot_1 = df_market_data_scaled_copy.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="clusters_4",
    hover_cols='coin_id',
    width=450
)

cluster_plot_1
```

---

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
market_data_pca = pca.fit_transform(df_market_data_scaled)

# Review the first 5 rows
market_data_pca[:5]
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
pca.explained_variance_ratio_
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** The total explained variance is about 89%

```python
# Create a new DataFrame with the PCA data.
df_pca = pd.DataFrame(market_data_pca, columns=["PCA1", "PCA2", "PCA3"])

# Set the coinid column as index
df_pca.index = df_market_data.index

# Display sample data
df_pca.head(10)
```

```python
df_pca.columns
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a a list to store inertia values and the values of k
inertia = []
k = list(range(1, 11))

# Append the value of the computed inertia from the `inertia_` attribute of teh KMeans model instance
for i in k:
    k_model = KMeans(n_clusters=i, random_state=1)
    k_model.fit(df_pca)
    inertia.append(k_model.inertia_)

# Define a DataFrame to hold the values for k and the corresponding inertia
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)

# Review the DataFrame
df_elbow.head()
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
# Plot the Elbow Curve
elbow2 = df_elbow.hvplot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve", 
    xticks=k
)
elbow2
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** I think the best value is `k=4`


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** Now the best value seems to be k=4 quite convincingly, where maybe k=3 was better before. 

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
model = KMeans(n_clusters=4, random_state=0)

# Fit the model
model.fit(df_pca)

# Make predictions
k_4 = model.predict(df_pca)

# Create a copy of the PCA DataFrame
df_pca_predictions = df_pca.copy()

# Add a class column with the labels
df_pca_predictions["clusters"] = k_4
df_pca_predictions.head()
```

```python
df_pca_predictions.columns
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.

cluster_plot_2 = df_pca_predictions.hvplot.scatter(
    x="PCA1",
    y="PCA2",
    by="clusters",
    hover_cols='coin_id',
    width=450
)
cluster_plot_2 
```
